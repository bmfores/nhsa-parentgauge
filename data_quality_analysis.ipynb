## VARIABLE FORMAT ANALYSIS

In our analysis of the DataFrame, most variables are stored as "object" data types, which can hold different types of information like text, numbers, or complex data. However, using "object" data types may have two main concerns. Firstly, it can use more memory, especially when dealing with large amounts of data or if the variables contain long text or complex information. This can impact the efficiency of our analysis and the amount of memory required to store the data. Secondly, performing calculations or operations on "object" data types may be slower compared to using specific data types. The flexibility of "object" data types means that extra processing is needed to handle different types of data within the same variable, which can slow down the analysis.

To address these concerns and improve performance, it is recommended to use specific data types that match the nature of the information in the variables. For example, using "text" data types for text data, "numbers" data types for numerical data, or "dates" data types for date-related information.
```python
# Define the styles for font colors
green_style = Style(color="green")
yellow_style = Style(color="yellow")
orange_style = Style(color="yellow")
red_style = Style(color="red")

# Counting total number of rows
row_count = df.shape[0]

# Create a Rich Table for missing value analysis
table = Table(title="Table of Missing Values")

table.add_column("Column", justify="left", style="bold cyan")
table.add_column("Missing Count", justify="right", style="bold magenta")
table.add_column("Missing Percentage", justify="right", style="bold green")

missing_count = df.isnull().sum()
total_count = df.shape[0]
missing_percentage = (missing_count / total_count) * 100

for column in df.columns:
    # Set the appropriate style based on the missing percentage
    if missing_percentage[column] <= 1:
        font_style = green_style
    elif missing_percentage[column] <= 25:
        font_style = yellow_style
    elif missing_percentage[column] <= 35:
        font_style = yellow_style
    else:
        font_style = red_style

    table.add_row(column, str(missing_count[column]), f"{missing_percentage[column]:.2f}%", style=font_style)

# Display the total row count
console = Console()
console.print(f"[underline]MISSING VALUE ANALYSIS[underline]")
console.print(f"Total number of rows: [bold]{row_count}[/bold]")

# Count the rows with different font colors
green_count = sum(row.style == green_style for row in table.rows)
yellow_count = sum(row.style == yellow_style for row in table.rows)
red_count = sum(row.style == red_style for row in table.rows)

# Print the counts
console.print(f"Variables with less than one percent (mostly complete) missing values: {green_count}")
console.print(f"Variables with less than thirty-five percent missing values: {yellow_count}")
console.print(f"Variables with over thirty-five percent percent missing values: {red_count}\n")

# Display the missing value analysis table
console.print(table)
```

```python
# Select columns starting with 'Q'
q_columns = [col for col in df.columns if col.startswith('Q')]

# Make a copy of the DataFrame to avoid modifying the original
#subset = filter_dataframe(df, conditions)

# Initialize a dictionary to store the breakdown results
breakdown_dict = {}

# Iterate over each 'Q' variable and calculate the breakdown
# Iterate over each 'Q' variable and calculate the breakdown
for column in q_columns:
    variable_conditions = conditions.get(column, {})  # Get the conditions specific to the current variable
    subset = df.copy()  # Make a copy of the DataFrame to avoid modifying the original

    # Apply the conditions for the current variable
    for condition_name, condition_data in variable_conditions.items():
        condition = condition_data['condition']
        operator = condition_data['operator']
        value = condition_data['value']
        subset = subset.loc[subset[condition].apply(lambda x: getattr(x, operator)(value))]

    values_count = subset[column].value_counts()
    total_count = values_count.sum()
    percentage = (values_count / total_count) * 100
    breakdown = pd.DataFrame({
        'Value': values_count.index,
        'Count': values_count.values,
        'Percentage': percentage.values
    })
    
    # Calculate the count and percentage for "Answered" and "Unanswered"
    answered_count = breakdown[breakdown['Value'] > 0]['Count'].sum()
    unanswered_count = breakdown[breakdown['Value'] < 0]['Count'].sum()
    total_count = breakdown['Count'].sum()
    answered_percentage = (answered_count / total_count) * 100
    unanswered_percentage = (unanswered_count / total_count) * 100
    
    # Create a new DataFrame for "Answered" and "Unanswered" rows
    answered_row = pd.DataFrame({'Value': ['Answered'], 'Count': [answered_count], 'Percentage': [answered_percentage]})
    unanswered_row = pd.DataFrame({'Value': ['Unanswered'], 'Count': [unanswered_count], 'Percentage': [unanswered_percentage]})

    # Replace the value strings with the corresponding context information
    breakdown['Value'] = breakdown['Value'].apply(lambda x: context.get(x, x))

    # Concatenate the new rows with the existing breakdown DataFrame
    breakdown = pd.concat([answered_row, unanswered_row, breakdown], ignore_index=True)

    breakdown_dict[column] = breakdown
    
# Generate the rich table for each variable
console = Console()

# Set the expand property to True
Table.expand = False

# Define the styles for the "Answered" and "Unanswered" rows
answered_style = Style(color="green")
unanswered_style = Style(color="red")

for column, breakdown in breakdown_dict.items():
    table = Table(show_header=True, header_style="bold magenta")
    table.add_column("Value")
    table.add_column("Count")
    table.add_column("Percentage")

    # Add context information for variable names
    variable_context = context.get(column, "")

    for _, row in breakdown.iterrows():
        value = row["Value"]
        count = row["Count"]
        percentage = row["Percentage"]

        if value == "Answered":
            style = answered_style
        elif value == "Unanswered":
            style = unanswered_style
        else:
            style = None

        # Add the row with the specified style
        table.add_row(
            Text(str(value), style=style),
            Text(str(count), style=style),
            Text("{:.2f}%".format(percentage), style=style),
        )

    console.print(f"Variable: [bold cyan]{column}[/bold cyan] - {variable_context}")
    console.print(table)
    console.print()
```
